```python
import matplotlib.pyplot as plt
from ipywidgets import *
```

```python
import numpy as np
```

```python
def f2nthp(n):
    plt.xlabel('x', fontsize=16)
    plt.ylabel(r'y=x$^\mathrm{%s}$'%(n), fontsize=16)
    a = np.arange(1, 10, 0.1)
    b = a ** n
    plt.plot(a, b, 'y.')
```

```python
f2nthp(2)
```

```python
interact(f2nthp, n=(-2.5,2.5,0.5))
```
